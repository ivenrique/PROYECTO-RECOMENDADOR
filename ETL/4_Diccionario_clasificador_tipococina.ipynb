```python
# Abrimos la lista type_cocina

ruta_archivo = "Sources/type_cocina_list"

with open(ruta_archivo, "rb") as archivo:
    type_cocina_list = pickle.load(archivo)
    
print(type_cocina_list)
```

```python
#Creamos un diccionario para clasificarlas guiandonos por el csv anterior
dictionary = {}
for i in range(130):
    value = input(f"{type_cocina_list[i]} :Ingrese el valor correspondiente: ")
    dictionary[type_cocina_list[i]] = value

print("\nDiccionario resultante:")
for key, value in dictionary.items():
    print(f"{key}: {value}")
```

```python
dictionary["Jamaicana"]="Caribeña"
```

```python
#Guardo el diccionario en un archivo pickle
# Especifica el nombre del archivo pickle y la carpeta de destino
nombre_archivo = 'dic_resumen_tipo cocina.pickle'
carpeta_destino = 'Sources'

ruta_archivo = os.path.join(carpeta_destino, nombre_archivo)

# Guarda el archivo en modo de escritura binaria (wb)
with open(ruta_archivo, 'wb') as archivo:
    # Utiliza la función dump de pickle para guardar el objeto en el archivo
    pickle.dump(dictionary, archivo)
```
