```python
#se quitan los NaN al diccionario
# Crear un nuevo diccionario sin los valores NaN
otros_tipo_cocina_sin_nan = {key: value for key, value in otros_tipo_cocina.items() if value!="nan"}
print(otros_tipo_cocina_sin_nan)
```

```python
#En este codigo se le pasa el diccionario clasificador a la columna tipos de cocina_list 
#para agrupar los tipos de cocina

columna_lista_tiposcocina= df_Rest_Tipo_Cocina["tipo_cocina_list"]

cocina_list_update = []

for i, j in enumerate(columna_lista_tiposcocina):
    #print(i, j)
    list_brief = []
    for x in j:        
        for key, value in dictionary.items():
            valor_a_comparar= x
            if valor_a_comparar == key:
                valor_a_comparar = value
                #print(valor_a_comparar)
                list_brief.append(valor_a_comparar)
                #print(f"lista:{list_brief}")
    cocina_list_update.append(list_brief)       

print(cocina_list_update[0:200])#hasta 200 para que se vea el ejemplo
```

```python
#Ahora se reduce el tamaño de la lista a un tipo d cocina que es el que mas se repite en la lista

#funcion para determinar el valor que mas se repite en una lista
def valor_mas_comun(sublista):
    if "Española" in sublista:
        return "Española"
    elif "Latina" in sublista:
        return "Latina"
    elif "Italiana" in sublista:
        return "Italiana"
    elif "China" in sublista:
        return "China"
    elif "Japonesa" in sublista:
        return "Japonesa"
    elif "Caribeña" in sublista:
        return "Caribeña"
    elif "Eur Occidental" in sublista:
        return "Eur Occidental"
    elif "America Sur" in sublista:
        return "America Sur"
    elif "India" in sublista:
        return "India"
    elif "Asiatica" in sublista:
        return "Asiatica"
    contador = Counter(sublista)
    valor, frecuencia = contador.most_common(1)[0]#el cero accede al primer elemento de la lista
    return valor

#Aquí está cómo funciona:
#contador.most_common(1) devuelve una lista de tuplas donde cada tupla contiene un 
#valor y su frecuencia correspondiente. El argumento 1 indica que solo quieres el 
#valor más común y su frecuencia.
#[0] se utiliza para acceder al primer elemento de esa lista de tuplas, que contiene el valor más
#común y su frecuencia.
#valor, frecuencia se desempaqueta la tupla en dos variables separadas.
#Entonces, al final de esa línea, valor contendrá el valor más común y frecuencia contendrá su frecuencia 
#en el contador.
#Este enfoque es muy útil cuando trabajas con el módulo collections.Counter para obtener información
#sobre los elementos más comunes en una secuencia.
    
```

```python
#se aplica la funcion para crear una columna con solo un tipo de cocina
cocina=cocina_list_update

tipo_cocina_uno = [valor_mas_comun(sublista) for sublista in cocina]

print(tipo_cocina_uno[0:10])#ejemplo de como se ve
```
