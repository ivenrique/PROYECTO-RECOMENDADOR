```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import math
from time import sleep
import pickle
```

```python
df=pd.read_csv("Sources\Trip_Rest_Madrid.csv")
```

```python
df["Direccion"][0]
```

```python
#Trasnformacion de las direcciones(limpieza..quitar comas, etc) para que funcione la API
direcciones_list=[]
for i,j in enumerate(df["Direccion"]):
    #print(i, j)
    if isinstance(j, str):#hay que poner esta condiccion porque hay valores que no son str
        a= j.replace("[","").replace("]","").replace(",","").replace("'","")
        direcciones_list.append(a.replace('"','').replace("  "," "))
    else:
        direcciones_list.append("No disponible")
print(direcciones_list[0:100])
```

#### APLICACION DE LA API DE GEOLOCALIZACION

```python
#Se crea un objeto Nominatim
geo=Nominatim(user_agent="MyApp")
```

```python
%%time
#Este codigo tarda 3hrs su ejecucion!!!!!!!!!!!!!!!!!!!!!!!!!!!!

latitud_list=[]
longitud_list=[]
c=0
for i in direcciones_list:
    c=c+1
    print(f"Falta {len(direcciones_list)- c}")
    try:
        address= geo.geocode(i, timeout=3)
        latitud_list.append(address.latitude)
        longitud_list.append(address.longitude)
        print(address.latitude,address.longitude)
        sleep(1)
    except:
        nan= np.nan
        latitud_list.append(nan)
        longitud_list.append(nan)
        print("NaN")
```
